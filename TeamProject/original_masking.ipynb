```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original = "/Users/yuna/Homework/BrailleBlockOriginal.jpeg"

origin = cv2.imread(original, 1)

height = 1920
width = 1080

# original의 크기를 영상에 맞추기
orig = cv2.resize(origin, (width, height), interpolation=cv2.INTER_LINEAR)

#original를 numpy 배열로 변환
orig = np.array(orig, dtype=np.uint8)

# 입력된 프레임을 HSV 형식으로 변환
hsv_frame = cv2.cvtColor(orig, cv2.COLOR_BGR2HSV)
```

```python
def show_img(image, title="image"):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
    
show_img(orig)
```
